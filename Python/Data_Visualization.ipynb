# 📊 Data Visualization — Superstore Sales
**Stack:** Matplotlib, Seaborn, Plotly

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
import plotly.express as px
sns.set_theme(style='whitegrid'); plt.rcParams['figure.figsize']=(9,5)
df = pd.read_csv('../Dataset/Cleaned_Data.csv', parse_dates=['order_date'])
df.shape
```

## 1. Histogram — Sales Distribution

```python
plt.hist(df['sales'], bins=60); plt.xlim(0,2000)
plt.title('Sales Distribution'); plt.xlabel('Sales ($)'); plt.ylabel('Frequency'); plt.show()
```

*Right-skewed: most orders are small; a few large B2B orders form a long tail.*

## 2. Boxplot — Sales by Category

```python
sns.boxplot(data=df, x='category', y='sales'); plt.ylim(0,1500)
plt.title('Sales by Category'); plt.show()
```

## 3. Bar Chart — Sales by Sub-Category

```python
s = df.groupby('sub_category')['sales'].sum().sort_values()
s.plot(kind='barh'); plt.title('Total Sales by Sub-Category'); plt.xlabel('Sales ($)'); plt.tight_layout(); plt.show()
```

## 4. Pie Chart — Revenue Share by Segment

```python
seg = df.groupby('segment')['sales'].sum()
plt.pie(seg, labels=seg.index, autopct='%1.1f%%', startangle=90)
plt.title('Revenue Share by Segment'); plt.show()
```

## 5. Heatmap — Correlation

```python
num=['sales','quantity','discount','profit','profit_margin_pct','unit_price','shipping_days']
sns.heatmap(df[num].corr(), annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Heatmap'); plt.tight_layout(); plt.show()
```

## 6. Scatter — Discount vs Profit

```python
sns.scatterplot(data=df.sample(2000,random_state=1), x='discount', y='profit',
                hue='category', alpha=.5)
plt.axhline(0,color='red',ls='--'); plt.title('Discount vs Profit'); plt.show()
```

*Higher discounts push more orders below the break-even line.*

## 7. Trend Chart — Monthly Sales & Profit

```python
t=df.groupby('order_year_month').agg(sales=('sales','sum'),profit=('profit','sum')).reset_index()
plt.plot(t['order_year_month'], t['sales'], label='Sales')
plt.plot(t['order_year_month'], t['profit'], label='Profit')
plt.xticks(t['order_year_month'][::3], rotation=45); plt.legend()
plt.title('Monthly Sales & Profit'); plt.tight_layout(); plt.show()
```

## 8. Interactive Charts (Plotly)

```python
fig = px.line(t, x='order_year_month', y='sales', title='Interactive Monthly Sales')
fig.show()
```

```python
cat = df.groupby('category').agg(sales=('sales','sum'),profit=('profit','sum')).reset_index()
fig = px.bar(cat, x='category', y=['sales','profit'], barmode='group',
             title='Sales vs Profit by Category'); fig.show()
```

```python
st = df.groupby('state')['sales'].sum().reset_index()
codes={'California':'CA','New York':'NY','Texas':'TX','Washington':'WA','Pennsylvania':'PA'}
# Full state->code mapping recommended for production choropleths
fig = px.treemap(df, path=['region','state'], values='sales', title='Sales by Region & State')
fig.show()
```

## 🧭 Business Interpretations
- **Concentration:** revenue is concentrated in a few sub-categories and top states — focus account management there.
- **Discount discipline:** the scatter shows discounting beyond ~25–30% routinely turns orders unprofitable.
- **Seasonality:** Q4 peaks justify inventory & staffing ramp-ups.
- **Mix shift:** grow Technology share to lift blended margin.
